방법1. 문자열 입력받기 -> 10진수 변환 -> 2진수 변환 -> 0b제거 (Fail)

```python
T = int(input())
for test_case in range(1, T + 1):
    n, num = input().split()  # 숫자와 문자(16진수의 A-F)를 입력받아야 하기 때문에 문자열로 입력받기 
    dec=int(num,base=16)      #16진수 num 10진수로 변환
    bin_result=bin(dec)       #10진수 dec 2진수로 변환
    result=bin_result[2:]     #이진수의 0b 부분을 슬라이싱하여 제거
    print("#"+str(test_case),str(result))  
```
